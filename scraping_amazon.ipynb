```python
# Importando as bibliotecas
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from time import sleep
import pandas as pd
```

```python
options = Options()
#options.add_argument('--headless')
#Definindo o tamanho da janela
options.add_argument('window-size=1500,850')

#Definindo o webdriver do chrome
navegador = webdriver.Chrome(options=options)

#Acessando o site
navegador.get('https://www.amazon.com.br/')

sleep(4)

#Encontrar a barra de pesquisa, inserir a palavra e clciar enter
input_place = navegador.find_element(By.XPATH, '//*[@id="twotabsearchtextbox"]')
input_place.send_keys('smartphone')
input_place.submit()

sleep(3)

#Clicar na opção com frete grátis
button_filter = navegador.find_element(By.XPATH, '//*[@id="p_n_free_shipping_eligible/19171733011"]/span/a/div[1]/label/i')
button_filter.click()

sleep(5)

#conteúdo da page
page_content = navegador.page_source

#Formato parser
pagina = BeautifulSoup(page_content, 'html.parser')

#Lista Vazia
smart_amazon = []

#Encontrar o elemento html com as informações de cada aparelho
page = pagina.findAll('div', attrs={'class': 'a-section a-spacing-base' })

#Função pra pegar as informações de cada aparelho e inserir na lista
for site in page:
    descricao = site.find('div', attrs={'class': 'a-section a-spacing-none a-spacing-top-small s-title-instructions-style'})
    avaliacao = site.find(attrs={'class': 'a-icon-alt'})
    preco = site.find(attrs={'class': 'a-price-whole'})
    decimal = site.find(attrs={'class': 'a-price-fraction'})
    if descricao is None:
        descricao = 'sem descrição'
    else:
        descricao = descricao.text
    if avaliacao is None:
        avaliacao = 'sem avaliação'
    else:
        avaliacao = avaliacao.text
    if preco is None and decimal is None: 
        preco = 'sem preço'
    else:
        preco = preco.text+decimal.text
    smart_amazon.append([descricao, avaliacao, preco])

#Transformar em um dataframe
dados = pd.DataFrame(smart_amazon, columns=['descrição', 'avaliação', 'preço'])

dados

```

```python
dados.to_csv("smart_amazon.csv", index=False)
```
